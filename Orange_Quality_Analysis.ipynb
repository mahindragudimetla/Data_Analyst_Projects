#Project-7:-Orange Quality Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/Orange Quality Data.csv")
```

```python
#1) Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?
correlation=data['Size (cm)'].corr(data['Weight (g)'])
print(correlation)
plt.figure(figsize=(13,6))
sns.barplot(x='Size (cm)',y='Weight (g)',data=data)
plt.xlabel('Size (cm)')
plt.ylabel('Weight (g)')
plt.title('Correlation between Size and Weight')
plt.show()
```

```python
#2) How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity?
correlation=data['Brix (Sweetness)'].corr(data['pH (Acidity)'])
print(correlation)
plt.figure(figsize=(25,6))
sns.barplot(x='Brix (Sweetness)',y='pH (Acidity)',data=data)
plt.xlabel('Brix (Sweetness)')
plt.ylabel('pH (Acidity)')
plt.title('Correlation between Brix and pH')
plt.show()
```

```python
#3) How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?
correlation=data['Softness (1-5)'].corr(data['Ripeness (1-5)'])
print(correlation)
plt.figure(figsize=(13,6))
sns.barplot(x='Softness (1-5)',y='Ripeness (1-5)',data=data)
plt.xlabel('Softness (1-5)')
plt.ylabel('Ripeness (1-5)')
plt.title('Correlation between Softness and Ripeness')
plt.show()
```
