#Project-3:-Social Media Sentiments Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/sentimentdataset.csv")
```

```python
#1) What is the overall distribution of sentiments (positive, negative, neutral) across the posts?
sentiment_counts = data['Sentiment'].value_counts()
print(sentiment_counts)

plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=data)
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Number of Posts')
plt.show()
```

```python
#2) Which platform generates the most engagement in terms of likes and retweets?
platform_engagement = data.groupby('Platform')[['Likes', 'Retweets']].sum() # Changed tuple to list
print(platform_engagement)
sns.barplot(x=platform_engagement.index, y=platform_engagement['Likes'], label='Likes')
sns.barplot(x=platform_engagement.index, y=platform_engagement['Retweets'], label='Retweets')
plt.title('Engagement by Platform')
plt.xlabel('Platform')
plt.ylabel('Engagement')
plt.legend()
plt.show()
```

```python
#3) Which users have the highest average engagement (likes and retweets per post)?
avg_engagement=data.groupby('User')[['Likes','Retweets']].mean()
print(avg_engagement)
```
