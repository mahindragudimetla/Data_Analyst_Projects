```python
#Now as we can see the dataset has null values so lets use the data cleaning to replace the null values
director_mode=data["director"].mode()[0]
data["director"].fillna(director_mode,inplace=True)
```

```python
#now similarly let replace the cast, country, date_added, rating , duration
```

```python
cast_mode=data["cast"].mode()[0]

data["cast"].fillna(cast_mode,inplace=True)
```

```python
country_mode=data["country"].mode()[0]
data["country"].fillna(country_mode,inplace=True)
```

```python
date_mode=data["date_added"].mode()[0]
data["date_added"].fillna(date_mode,inplace=True)
```

```python
rating_mode=data["rating"].mode()[0]
data["rating"].fillna(rating_mode,inplace=True)
```

```python
duration_mode=data["duration"].mode()[0]
data["duration"].fillna(duration_mode,inplace=True)
```

```python
data.isna().sum()

#as we can see we have replaced the missing values(null values)
```
