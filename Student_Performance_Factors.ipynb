```python
"""DATA CLEANING:-
First lets see where we need to clean the data

Here we have 0's in the Teacher_Quality, Learning_Disabilities, and Distance_From_Home
Let's use the data cleaning to remove the null(0's) values from the dataset
"""
```

```python
data["Teacher_Quality"].head()
```

```python
#Removing null values from Teacher_Quality
print("Now let's Find the highest repeated value using mode")
Teacher_Quality_mode=data["Teacher_Quality"].mode()[0]
# Here we use the [0] because we need to make changes into the "Teacher_Quality" coulmn
#We do not need to use the [0] for everything we only use it for the mode
print("The highest repetead value is:-",Teacher_Quality_mode)
```

```python
data["Teacher_Quality"].value_counts()
#by using this we can see the amount of times each value is repeated
```

```python
#Now that we have the highest repeated value in the Teacher_Quality lets use it to fill the null values

data["Teacher_Quality"].fillna(Teacher_Quality_mode,inplace=True)

#as we can see the Teacher_Quality null values have been removed an replaced with real values
# based on this lets do the others as well
```

```python
# lets do the data cleaning for the Parental_Education_Level

print("Now let's Find the highest repeated value using mode")
Parental_Edu_mode=data["Parental_Education_Level"].mode()[0]
# Here we use the [0] because we need to make changes into the "Parental_Education_Level" coulmn
#We do not need to use the [0] for everything we only use it for the mode
print("The highest repetead value is:-",Parental_Edu_mode)
```

```python
data["Parental_Education_Level"].value_counts()
```

```python
data["Parental_Education_Level"].fillna(Parental_Edu_mode, inplace=True)

#now that even Parental_Education_Level is completed let's move onto the final one Distance_from_Home
```

```python
# now finally we can proceed onto doing data cleaning for the Distance_from_Home

print(" let's Find the highest repeated value using mode")
Distance_mode=data["Distance_from_Home"].mode()[0]
# Here we use the [0] because we need to make changes into the "Parental_Education_Level" coulmn
#We do not need to use the [0] for everything we only use it for the mode
print("The highest repetead value is:-",Distance_mode)
```

```python
data["Distance_from_Home"].value_counts()
```

```python
data["Distance_from_Home"].fillna(Distance_mode, inplace=True)
data.isna().sum()
```

#Tasks



```python
#1) How does the number of hours studied per week correlate with the final exam score?
correlation = data['Hours_Studied'].corr(data['Exam_Score'])
print(f"The correlation between hours studied per week and the final exam score is: {correlation}")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=data)
plt.title('Correlation between Hours Studied and Final Exam Score')
plt.xlabel('Hours Studied per Week')
plt.ylabel('Final Exam Score')
plt.show()
```

```python
#2) Is there a significant relationship between class attendance percentage and exam scores?
correlation = data['Attendance'].corr(data['Exam_Score'])
print(correlation)
plt.figure(figsize=(8, 6))
sns.lineplot(x='Attendance', y='Exam_Score', data=data)
plt.title('Correlation between Attendance and Final Exam Score')
plt.xlabel('Attendance Percentage')
plt.ylabel('Final Exam Score')
```

```python
#3) Does the average number of sleep hours per night influence exam performance?
correlation = data['Sleep_Hours'].corr(data['Exam_Score'])
print(correlation)
plt.figure(figsize=(8, 6))
sns.barplot(x='Sleep_Hours', y='Exam_Score', data=data)
plt.title('Correlation between Sleep Hours and Final Exam Score')
plt.xlabel('Average Sleep Hours per Night')
```

```python
#4) How does family income and parental education level impact the student's exam scores?
plt.figure(figsize=(10, 6))
sns.barplot(x='Family_Income', y='Exam_Score', hue='Parental_Education_Level', data=data)
plt.title('Impact of Family Income and Parental Education on Exam Scores')
plt.xlabel('Family Income')
plt.ylabel('Exam Score')
plt.show()
```

```python
#5) Do private school students perform better in exams than public school students?
plt.figure(figsize=(8, 6))
sns.barplot(x='School_Type', y='Exam_Score', data=data)
plt.title('Comparison of Exam Scores between Private and Public Schools')
plt.xlabel('School Type')
plt.ylabel('Exam Score')
```
